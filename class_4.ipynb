```python
# when is the most products baught
# if an expensive product is baought it is baught at the same time
# dettermine the number of varients
# design an experiment
import matplotlib.pyplot as plt
```

```python

dfr = df.groupby('category_code')['price'].sum().reset_index()
dfr.sort_values('price', axis=0, ascending=True, inplace=True)
dfr=dfr.drop(dfr[dfr['price']==0].index,axis=0)

fig, ax = plt.subplots(figsize=(5, 30))
sns.barplot(data=dfr, x='price', y='category_code')
```
